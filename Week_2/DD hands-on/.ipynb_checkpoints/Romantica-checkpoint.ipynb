In the resource estimation for QPE, they mentioned that with Iterative QPE, you only need one ancillary qubit. Here's the code for QPE and IQPE with the unitary being measured being the phase shift. 

Play around with it, and figure out what they meant by you only need one qubit. No need to explain how and why they do it, but what seems to change when you look at the two circuits?

Hint: There are two registers for QPE normally. Which register are they changing? (Ask if you need explanation)

```python
import pennylane as qml
import numpy as np

def U(wires):
    return qml.PhaseShift(2 * np.pi / 5, wires=wires)
```

#### QPE code

```python
dev = qml.device("default.qubit")

@qml.qnode(dev)
def circuit_qpe(estimation_wires):
    # initialize to state |1>
    qml.PauliX(wires=0)

    for wire in estimation_wires:
        qml.Hadamard(wires=wire)

    qml.ControlledSequence(U(wires=0), control=estimation_wires)

    qml.adjoint(qml.QFT)(wires=estimation_wires)

    return qml.probs(wires=estimation_wires)
```

```python
estimation_wires = range(1, 5)

print(qml.draw_mpl(circuit_qpe)(estimation_wires))
```

```python
import matplotlib.pyplot as plt

estimation_wires = range(1, 5)

results = circuit_qpe(estimation_wires)

bit_strings = [f"0.{x:0{len(estimation_wires)}b}" for x in range(len(results))]

plt.bar(bit_strings, results)
plt.xlabel("phase")
plt.ylabel("probability")
plt.xticks(rotation="vertical")
plt.subplots_adjust(bottom=0.3)

plt.show()
```

#### IQPE code

```python
dev = qml.device("default.qubit", shots=10)

@qml.qnode(dev)
def circuit():

  # Initial state
  qml.X(0)

  # Iterative QPE
  measurements = qml.iterative_qpe(U(wires=0), ancilla=1, iters=4)

  return qml.sample(measurements)
```

```python
result = circuit()
result
```
